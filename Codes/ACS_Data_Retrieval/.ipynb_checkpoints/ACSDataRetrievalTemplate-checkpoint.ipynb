### General Overview
In order to retrieve census data for the ACS, my general steps were:
1. Identify the variables of interest by referring to this page: https://www.census.gov/programs-surveys/acs/guidance/which-data-tool/table-ids-explained.html
    - In particular, I looked at element 1, 2, and 3 tables
    - **Element 1** For this project, I chose to focus on the B table type of which contains the "Most detailed estimates on all topics for all geographies"
    - **Element 2** I chose to focus on subject numbers 02 (Race),03 (Hispanic or Latino Origin),15 (Educational Attainment),19 (Income). **Please feel free** to add more 
    - **Element 3** I went with 001 as I wanted to look at general values. Any other 2 or 3 digit number IDs contained a more granular look at the subject of interest. As an example for the race subject number the 009 gave data on `Universe:  Black or African American alone or in combination with one or more other races` As such, 001 was chosen.
 2. Once I had the variables of interest down, then I went ahead and retrieved the data of interest via an API call using the census python package (see here for more info: https://pypi.org/project/census/).
 3. I then  reshaped the data to be in a long format for ease of use when analyzing the data.
 4. Lastly, I added a type (ie. Race) and subtype (ie. White) columns for readability

### Loading in table shells : "They contain the line number, description of the data, and table ID" - Census

This is how we'll be able to track down the variables of interest

```python
shell = pd.read_excel("https://www2.census.gov/programs-surveys/acs/summary_file/2019/documentation/user_tools/ACS2019_Table_Shells.xlsx")
```

## Retrieving Element 2 Table From Census Website
##### Contains detailed table type contains the most detailed estimates on all topics for all geographies

```python
page = requests.get("https://www.census.gov/programs-surveys/acs/guidance/which-data-tool/table-ids-explained.html").text
# Element 2 Table: The next 2 characters identify the subject of the table. 
subject2Table = soup(page).find_all("div",{"class":"uscb-accordion"})[1]

def retrieveText(tags):
    """
    Retrieve all text from each row in subject 2 table.
    Split by end of line and then filter out any empty strings.
    End result should be a list with 4 element (ie. [01, 'Age;Sex', 17, 'Poverty Status'])
    """
    text = tags.text
    listText = text.split('\n')
    notEmpty = lambda x: x!=''
    return list(filter(notEmpty, listText))

def definitionTable(table):
    """
    Takes filtered table converts them into a pandas dataframe.
    Seperates each into 2 sections as they both reference to the same columns
    Then stacks 2nd section to the bottom of the first section.
    """
    df = pd.DataFrame(data= table[1:])
    firstSection  ={0:"Subject Number", 1:"Subject Name"}
    secondSection ={2:"Subject Number", 3:"Subject Name"}
    return (df[[0,1]]
            .rename(columns=firstSection)
            .append(df[[2,3]]
                .rename(columns=secondSection))
           )

filteredTable  = list(map(retrieveText, subject2Table.find_all("tr")))
subjectMap = definitionTable(filteredTable).iloc[:-1]
subjectMap
```

## Function to retrieve variables of interest.
Elements referenced. See here: https://www.census.gov/programs-surveys/acs/guidance/which-data-tool/table-ids-explained.html

```python

def mainTable(subjNum):
    """
    Retrieves main table consisting of subject number of interest.
        - See Element 2 for reference 
    Keeps only rows with table IDs equal to 6 as further detailed variables are
        not of interest. See Element 4 and 5 for reference:
    """
    if type(subjNum) in [int, float]:
        subjNum = str(int(subjNum))
        subjNum =  "0"+subjNum if len(subjNum)==1 else subjNum
    queryMatch = shell['Table ID'].str.startswith("B"+ subjNum).fillna(False) 
    generalStats = shell['Table ID'].fillna("").apply(len) == 6
    return shell[(queryMatch) & (generalStats)]

def subjNames(df):
    """
    Returns dataframe with the main subject types.
    Generally you want to keep 001 as it gives a general view of the data.
    Any other subject types contain a more granular look at the subject of interest. 
          -  See Element 3 for reference.
    """
    tableIDs = df['Table ID'].unique()
    return shell[shell['Data Release'].notnull() & df['Table ID'].isin(tableIDs)]

def table(df,tableID):
    """
    Return dataframe with specific table of interest.
    """
    return df[df["Table ID"]== tableID].applymap(lambda x: np.nan if type(x)==str and x.isspace() else x)
```

## Table ID B02001: RACE

```python
race = mainTable("02")
race = table(race, "B02001")
race 
```

## Table ID B15003: EDUCATIONAL ATTAINMENT FOR THE POPULATION 25 YEARS AND OVER

```python
education = mainTable("15")
education = table(education, "B15003")
education
```

## Table ID B19101: FAMILY INCOME IN THE PAST 12 MONTHS (IN 2019 INFLATION-ADJUSTED DOLLARS)

```python
familyHHIncome = mainTable("19")
familyHHIncome = table(familyHHIncome, "B19101")
familyHHIncome
```

## Table ID B03002: HISPANIC OR LATINO ORIGIN BY RACE

```python
isHispanic = mainTable("03")
isHispanic = table(isHispanic, 'B03002')
isHispanic
```

## Table ID B25063: GROSS RENT

```python
grossRent = mainTable("25")
grossRent = table(grossRent,"B25063")
grossRent
```

```python
#Include invariants, i.e, variables of interest, acs, etc.
acs_year = 2019
#set up API key
key ="650e1f456fbc088923173abf25a47c07ee49d021" # sign up for a key here: https://api.census.gov/data/key_signup.html
```

## Defining our fields of interest using the Unique ID as our reference points. Please note that an extra character, "E", is added here as it pertains to the Estimate values. Lastly, the line is used as query to retrieve the variables of interest

```python
(race.query("1< Line < 9")['UniqueID'] + "E").values
```

```python
#create list of desired vars 
race_vars = list((race.query("1< Line < 9")['UniqueID'] + "E").values)
education_vars = list((education.query("1< Line")['UniqueID'] + "E").values)
family_hh_income = list((familyHHIncome.query("1 < Line")['UniqueID'] + "E").values)
latino_population = ["B03002_002E", "B03002_012E"]
rent_vars = list((grossRent.query("2 < Line < 27")['UniqueID'] + "E").values)

#merging variables into columns and then converting into a tuple
census_var_list = race_vars + education_vars + family_hh_income + latino_population+ rent_vars
```

### Calling API at the census tract level. We were able to retrieve the corresponding state and city FIPS codes via this link:
- https://www2.census.gov/programs-surveys/decennial/2010/partners/pdf/FIPS_StateCounty_Code.pdf

```python
#downloading the census data
c = Census(key=key, year = acs_year)
census_data_census_tract = c.acs5.get(fields = census_var_list, 
                                      geo = {'for': 'block group:*' , 
                                             'in': 'state:{} county:{}'.format('37', '063')}
                                     )
census_tracts = pd.DataFrame(census_data_census_tract)
census_tracts.head(2)
```

## Reshaping dataframe from wide format (many columns) to a long format

```python
census_tracts = census_tracts.melt( id_vars=['state', 'county', 'tract','block group'],
       value_vars=census_var_list)
census_tracts.head(2)
```

## Adding Type (ie. Race) and SubType (ie. White) columns for readability

```python
demoType = {'B02001':"Race", 'B15003':"Education",
            'B19101':"Family Household Income","B25063":"Gross Rent",
            'B03002':"Hispanic/Latino"}
# indexing up to 6 character -> the length of the table ID
census_tracts['Type'] = census_tracts['variable'].apply(lambda x: demoType[x[:6]])
census_tracts.head(3)
```

Retrieving columns with the variables of interest in order to use their more  readable variable names 

```python
filterE = lambda x: x.replace("E","")
variables =list(map(filterE,census_var_list))

# dictionary {census unique ID  : readable variable name}
demoSubtype = shell.query("UniqueID in @variables").set_index('UniqueID')['Stub'].to_dict()
demoSubtype.values()
```

```python
census_tracts['Subtype'] = census_tracts['variable'].apply(lambda x: demoSubtype[x[:-1]])
```

## General EDA

```python
def getType(df,x):
    return df[df['Type'] == x]
```

```python
census_tracts['Type'].unique()
```

### Predominant Subtype Per Census Tract

```python
def getMax(df,subtype):
    data = (getType(df,subtype)
     .pivot_table(index='tract',
                  values='value',
                  columns='Subtype',
                  aggfunc='sum')
           )
    
    print(f"\t----Predominant {subtype} Value Counts----")
    display(data
     .apply(lambda x: x.idxmax(),1)
     .value_counts()
    )
    print(f"\n\t----Predominant {subtype} Absolute Count ----")
    display(data
     .apply(lambda x: x.max(),1)
     .to_frame().T
    )
```

```python
getMax(census_tracts, 'Race')
getMax(census_tracts, 'Education')
getMax(census_tracts, 'Family Household Income')
getMax(census_tracts, 'Hispanic/Latino')
getMax(census_tracts, 'Gross Rent')
```

```python
g = sns.catplot(x="tract",y="value", col="Subtype", col_wrap=3,
                data=getType(census_tracts,'Race'), ci=None, palette='pastel',
                kind="bar", height=3, aspect=1)
g.set_xticklabels([])
g.set_titles("{col_name}")
```
